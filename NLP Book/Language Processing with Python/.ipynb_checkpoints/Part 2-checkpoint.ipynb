<h3>Frequency Distribution</h3>

<p>How can we identify about the words in a text that are most informative about the topic and genre</p>

<p>Frequency Distribution tells up frequecy of each vocabulary in a text</p>
<p> It tells about the total number of word tokens in a text are distributed across the vocabulary items</p>
<p> This can be done using FreqDist(text_name)</p>

```python
freqdist = FreqDist(text1)
freqdist
voabulary1 = freqdist.keys()
voabulary1[:50]
```

```python
freqdist.plot(50, cumulative=True)
```

<p>Words that occurs only once are called hapaxes</p>

```python
freqdist.hapaxes()
```

<h6>Fine Grained Selection of Words</h6>

<p> It means selecting particular words that are specific characters long</p>

```python
# Here we are finding words that are greater than 15 characters

V = set(text1)
long_words = []
for W in V:
    if len(W)>15:
        long_words.append(W)
long_words
```

<h6> Collocations and Bigrams</p>

<p> A Collocations is a sequence of words that occurs together unusually often. Thus red wine is a collocation, whereas the wine is not. A characteristic of collocations is that
they are resistant to substitution with words that have similar senses; for example, maroon wine sounds very odd.</p>

<p> Bigrams are the words that occurs in pairs </p>

```python
print bigrams(['more','is','said','than','done'])
```

<p>In particular, we want to find bigrams that occur more often than we would expect based on the frequency of individual words.</p>

```python
text4.collocations()
```
